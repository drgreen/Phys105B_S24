### Basic Data Types

We will use a lot of numpy in this course. We import the numpy library as follows:

```python
import numpy as np
```

The basic object in numpy is the array. It combines the best features of numbers and lists in python. In essence, an array is a list of numbers that you can perform math operations on as if it is a single number. 

To see that value of arrays, it is helpful to start from the basic python data types. Numbers are usual of the time int (integer) or float (decimal). E.g. we can work with integers like this

```python
a=5
b=6
print(a+b,a*b,a**2)
```

or floating point numbers like this

```python
aa=5.
bb=6.14
print(aa+bb,aa*bb,aa**2)
```

Integers are automatically converted to floats if you used mixed operations:

```python
print(a*2,a*2.)
```

Some function can only take integers as input so you may need to convert floats to ints via

```python
print(int(aa),int(bb))
```

Numbers are distinct from words, which are called strings and appear as

```python
c='Hi'
d='Prof Green'
print(c+d)
print(c+' '+d)
```

A number can be a string too, so it can cause confusion

```python
print(1+1)
print('1'+'1')
```

The final common data-type is a Boolean variable (bool), which is just a true or false.

```python
bool1=True
bool2=False
```

Usually, we create bools by logical operations:

```python
print(1>0,0<1,2==3)
print(bool1 and bool2)
print(bool1 or bool2)
```

Without numpy, we can still group numbers together using a list:

```python
list1=[5,6]
```

Lists are fine, but math operations don't act on the numbers.

```python
list1*2
```

What is happening here is that rather than adding or multiplying the numbers in the list, it is making a bigger list by adding elements of the list together.

While we will see later how to loop over the elements of a list, we can skip the trouble using a numpy array in place of the list. We can convert a list to a numpy array as follows: 

```python
arr1=np.array([5,6])
print(arr1*2,arr1+1,arr1**2)
```

Notice that using the numpy array, basic math operations act on each element of the array seperately, giving an array with the results of the calculation as output.

**Remember** the entries in lists or numpy arrays are labeled by an integer starting at 0. So the first entry in a list is 0 and the second is 1, and so on:

```python
print(arr1[0])
print(arr1[1])
```

There are a few useful tool for quickly making a numpy array. To make an array that is an ordered list of integers:

The same function can be used to make a list of numbers between two endpoints with a specific step size

```python
np.arange(0,1,0.01)
```

#### Loops and Logic

For loops and while loops are very useful tools for doing anything repetative. The syntax of python is to use tabs to distinguish what is happening inside a loop versus before or after.

A for loop is built on the idea of just taking a list and running through each of the entries. You might want to do something with the entries along the way. For example, we could add all the numbers in a list

```python
sum=0
numlist=np.arange(7)
for item in numlist:
    print(item)
    sum+=item
print('this is the answer:',sum)
```

Rather than run over the items in your list, it is sometimes more useful to run over the index of the list. One way to do this is range:

```python
sum=0
numlist2=np.arange(7)*3+1
for i in range(len(numlist2)):
    print(i,numlist2[i])
    sum+=numlist2[i]
print('this is the answer:',sum)
```

A while loop is another strategy for looping over a set of operations. The key benefit of while is that it works well when you aren't sure how many times you need to run through the loop. While has a true or false statement statement defined at the beginning. The loop keeps running as long as the statement continues to be true.

```python
a=10
while a>=-8:
    print(a)
    a-=3
print('the loop is done: a=',a)
```

A common tool you may want to use is an if statement inside of the loop. If statements are also organzied by tabs. E.g. suppose I want to run through a list of numbers and print only the numbers between 5 and 10: 

```python
for item in numlist2:
    if item>=5 and item<=10:
        print(item)
```

If, else if and else can be used together to make decision trees

```python
for item in numlist2:
    if item<5:
        print('too small')
    elif item>=5 and item<=10:
        print(item)
    else:
        print('too big')
```

### Shapes of numpy arrays

While the most basic array is just a list, arrays can be made into more complex shapes. The simplest generalization is something like a matrix:

```python
arr2d=(np.linspace(1,9,9)).reshape(3,3)
print(arr2d)
print(arr2d.shape)
```

This works like a matrix in the sense that the individual entries are listed by 2 numbers, one for the row and one for the column

```python
print(arr2d[0,0],arr2d[0,1],arr2d[1,0])
```

We can isolate an entire row or column using the colon for one of the indices

```python
print(arr2d[0,:])
print(arr2d[:,0])
```

For any shape, it is important to remember that math operations work on each number individually. E.g. 

```python
arr2d*arr2d
```

This is just the square of every entry. If we want matrix multiplication, we need to specify

```python
np.dot(arr2d,arr2d)
```

We are not limited to arrays in the form of vectors and matrices. We can make them any dimensionality

```python
arr3d=(np.linspace(1,9,27)).reshape(3,3,3)
print(arr3d)
print(arr3d.shape)
```

```python
arr3d[0,1,2]
```

### Functions and Numpy operations

A function is a tool to take a bunch of operations you do regularly and have them stored to use at any time. It is again defined using the tab structure.

```python
def add(a,b):
    return a+b
```

```python
add(5,7)
```

### Figures

```python
import matplotlib.pyplot as plt
```

Python has a lot of flexibility in making figures. If you are just trying to make a simple figure quickly, plt.plot() is a good option:

```python
x_list = [0,1,2,3]
y_list= [0,2,4,9]
```

```python
plt.plot(x_list,y_list)
plt.show()
```

Any of the plotting functions we can use with numpy or other tools we have developed so far.

```python
time=np.linspace(0,2*np.pi,100)
xt={}
for i in range(3):
    xt[i]=np.cos(time*(i+1)**2)
```

```python
for i in range(3):
    plt.plot(time,xt[i])
plt.show()
```

The more powerful, but somewhat more involved, techniques tend to define figures using the "axes" framework. We can generate axes quickly and easily as follows:

```python
fig,ax=plt.subplots()
```

Here I just made a blanck figure. One important use of axes is that you can have a single "figure" with multiple axes

```python
fig2,ax2=plt.subplots(figsize=(10,8),nrows=2,ncols=3)
```

Now we make the figure by ploting uses the specific axes:

```python
fig,ax=plt.subplots()
ax.plot(x_list,y_list)
plt.show()
```

When we have multiple axes in a single figure, we call them like an array.

```python
fig2,ax2=plt.subplots(figsize=(10,8),nrows=2,ncols=3)
for i in range(3):
    for j in range(2):
        ytemp=np.cos(time*(i+1)**2+np.pi*j)
        ax2[j,i].plot(time,ytemp)
plt.show()
```

Most of the time, we can look up the rules for the kind of figure we want, when we need it. However, there are a few common adjustments that we make to figures that are worth mentioning up front.

1. Changing the limits of the x and/or y axes

```python
x=np.linspace(0,2*np.pi,300)
y1=np.cos(x*3)
y2=np.cos(x**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(0,2*np.pi)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlim(0,2*np.pi)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xlim(0,2*np.pi)
ax.set_ylim(-2,2)
plt.show()
```

2. Changing line colors / line styles - same syntax for ax.plot and plt.plot

```python
fig,ax=plt.subplots()
ax.plot(x,y1,color='darkgreen',linestyle='-.') # c or color does the same thing
ax.plot(x,y2,c='purple',ls='--',lw=3) # ls/lw or linestyle/linewidth are the same
ax.set_xlim(0,2*np.pi)
plt.show()
```

3. log scale on the x or y axis

```python
x=np.linspace(0,2*np.pi,300)
y1=np.cos(x*3)
y2=np.cos(x**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xscale('log')
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)
ax.set_xscale('log')
plt.show()
```

4. adding a legend

```python
fig,ax=plt.subplots()
ax.plot(x,y1,label='y1')
ax.plot(x,y2,label='y2')
ax.set_xscale('log')
ax.legend()
plt.show()
```
